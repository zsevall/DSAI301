```python
#zeynep_seval_2024660042_hw2

# 1a
def quadraticSolution(a,b,c):
    # Check if equation is quadratic
    if a == 0:
        print("The entered equation is not quadratic")
        return
# 1b
#how to find solutions of a quadratic equation:
#check webpage https://www.cuemath.com/algebra/discriminant/
    else:
      discriminant = b*b - 4*a*c
      # Case 1: Discriminant > 0 (Two real solutions)
      if discriminant > 0:
        print("Your equation have two real solution.")
        # Calculate roots
        sol1 = (-b + discriminant ** 0.5) / (2*a)
        sol2 = (-b - discriminant ** 0.5) / (2*a)
        print(f"Solution 1 : {sol1:.2f}")
        print(f"Solution 2 : {sol2:.2f}")

      # Case 2: Discriminant = 0 (One real solution)
      elif discriminant == 0:
        print("Your equation have one real solution.")
        # Calculate solution: -b / (2a)
        sol = -b / (2*a)
        print(f"Solution : {sol:.2f}")

      # Case 3: Discriminant < 0 (Two complex solutions)
      else:
        print("Your equation have two complex solution.")
        # Calculate real and imaginary parts
        real_part = -b / (2*a)
        imagine_part = ((-discriminant) ** 0.5) / (2*a)
        if real_part != 0:
          print(f"Solution 1: {real_part:.2f} + {imagine_part:.2f}i")
          print(f"Solution 2: {real_part:.2f} - {imagine_part:.2f}i")
        else:
          print(f"Solution 1: {imagine_part:.2f}i")
          print(f"Solution 2: -{imagine_part:.2f}i")

quadraticSolution (0,1,1)
quadraticSolution (1,2,5)
quadraticSolution (1,0,4)
quadraticSolution (2,1,-2)
quadraticSolution (1,-2,1)
```
